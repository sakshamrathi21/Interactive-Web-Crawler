```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/C H Traders/Downloads/data.csv")
name = df['Name']
data1 = df['Data1']
data2 = df['Data2']
```

```python
plt.plot(name, data1, marker = 'o', label = 'data 1')
plt.plot(name, data2, marker = 'o', label = 'data 2')
plt.legend()
plt.title("Line Plot")
plt.xlabel('Name')
plt.ylabel("Data")
```

```python
plt.figure(figsize=(10, 6))
bar1 = range(len(df['Name']))
width = 0.35
bar2 = [x + width for x in bar1]
plt.bar(bar1, data1, width= width, label = 'data 1')
plt.bar(bar2, data2, width = width, label = 'data 2')
plt.legend()
plt.xlabel('name')
plt.ylabel("data")
plt.title("bar plot")
```
