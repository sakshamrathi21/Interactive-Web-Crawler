```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Chiransh/Downloads/randomdata.csv")
```

```python
xval = df['Month']
yval_1 = df['Sales A']
yval_2 = df['Sales B'] 
```

```python
plt.plot(xval, yval_1, label = 'Sales A', marker = 'o')
plt.plot(xval, yval_2, label = 'Sales B', marker = 'o')
plt.legend()
plt.xlabel('Month')
plt.ylabel("Sales")
plt.title("Monthly sales")
```

```python
plt.figure(figsize=(10, 6))
bar1 = range(len(df['Month']))
width = 0.35
bar2 = [x + width for x in bar1]
plt.bar(bar1, yval_1, width= width, label = 'Sales A')
plt.bar(bar2, yval_2, width = width, label = 'Sales B')
plt.legend()
plt.xlabel('Month')
plt.ylabel("Sales")
plt.title("Monthly sales")
```
